# Learning Paradigms Overview

## Unsupervised Learning - Association


### Association Rule Mining Using FP Growth Algorithm

```python
#import libraries
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rules,apriori
```

```python
#get data
dataset = pd.read_csv("supermarket.csv")
dataset.head()
```

```python
dataset = dataset.replace('?',False)
dataset = dataset.replace('t',True)
```

```python
dataset.head()
```

```python
#One hot encoding for total 
one_hot_total = pd.get_dummies(dataset['total'])
dataset = pd.concat([dataset,one_hot_total],axis =1)
dataset = dataset.drop('total',axis =1)
dataset.head()
```

```python
dataset = dataset.replace(0,False)
dataset = dataset.replace(1,True)
dataset.head()
```

##### FP Growth 

```python
#obtaining frequent itemsets using FPGrowth
frequent_itemsets1 = fpgrowth(dataset, min_support=0.5, use_colnames=True)
```

```python
frequent_itemsets1
```

```python
rules_by_lift1 = association_rules(frequent_itemsets1,metric="lift",min_threshold=0.8)
rules_by_lift1.shape
```

```python
rules_by_lift1
```

```python
rules_by_conf1 = association_rules(frequent_itemsets1,metric="confidence",min_threshold=0.6)
rules_by_conf1
```

##### Using Apriori Method


```python
frequent_itemsets2 = apriori(dataset, min_support=0.5, use_colnames=True)
frequent_itemsets2
```

```python
rules_by_lift2 = association_rules(frequent_itemsets2,metric="lift",min_threshold=0.8)
rules_by_lift2.shape
```

```python
rules_by_lift2
```

```python
rules_by_conf2 = association_rules(frequent_itemsets2,metric="confidence",min_threshold=0.6)
rules_by_conf2.shape
```

```python
rules_by_conf2
```
